```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

# Importing data from sklearn

```python
iris_dataset = datasets.load_iris()
```

```python
iris_dataset
```

### Converting data to pandas dataframe

```python
x = pd.DataFrame(data= np.c_[iris_dataset['data'], iris_dataset['target']], columns= iris_dataset['feature_names'] + ['target'])
```

```python
x["petal length cat"] = pd.cut(x["petal length (cm)"], bins=[0, 1.39, 2, 5, 6, 7], labels=[1, 2, 3, 4, 5])
x["petal length cat"].hist()
```

```python
x["petal width cat"] = pd.cut(x["petal width (cm)"], bins=[0, 0.3, 1.9, 3], labels=[1, 2, 3])
x["petal width cat"].hist()
```

# Creating Train and Test sets

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)
for train_index, test_index in split.split(x, x["petal length cat"], x["petal width cat"]):
    train_set = x.loc[train_index]
    test_set = x.loc[test_index]
```

```python
test_set["petal length cat"].value_counts()
```

```python
test_set["petal width cat"].value_counts()
```

```python
for set_ in (train_set, test_set):
    set_.drop(["petal width cat", "petal length cat"], axis=1, inplace=True)
```

# Correlation Between Attributes

```python
corr_matrix = train_set.corr()
corr_matrix["target"].sort_values(ascending=False)
```

# Feature Engineering

```python
train_set["petal length per width"] = train_set["petal length (cm)"]/train_set["petal width (cm)"]
train_set["sepal length per width"] = train_set["sepal length (cm)"]/train_set["sepal width (cm)"]
train_set["petal length plus width"] = train_set["petal length (cm)"]+train_set["petal width (cm)"]
train_set["sepal length plus width"] = train_set["sepal length (cm)"]+train_set["sepal width (cm)"]
train_set["petal length plus width sepal length plus width"] = train_set["petal length (cm)"]+train_set["petal width (cm)"]+train_set["sepal length (cm)"]+train_set["sepal width (cm)"]
```

```python
corr_matrix = train_set.corr()
corr_matrix["target"].sort_values(ascending=False)
```
